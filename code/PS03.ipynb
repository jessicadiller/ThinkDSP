## 2) Compute the autocorrelation of BitCoin prices.

```python
def corrcoef(xs, ys):
    return numpy.corrcoef(xs, ys, ddof=0)[0, 1]

def serial_corr(wave, lag=1):
    N = len(wave)
    y1 = wave.ys[lag:]
    y2 = wave.ys[:N-lag]
    corr = corrcoef(y1, y2)
    return corr

def autocorr(wave):
    lags = range(len(wave.ys)//2)
    corrs = [serial_corr(wave, lag) for lag in lags]
    return lags, corrs
```

```python
nrows = 1625
df = pandas.read_csv('coindesk-bpi-USD-close.csv', nrows=nrows, parse_dates=[0])
ys = df.Close.values
wave = thinkdsp.Wave(ys, framerate=1)
lags, corrs = autocorr(wave)
thinkplot.plot(lags, corrs)
```

## Does the autocorrelation function decay exponentially? Is there evidence of periodic behavior?

### Yes this autocorrelation function decays exponenetially and there is evidence of periodic behavior as you can see because it peaks a couple times as the curve is generally decreasing. ZThe peaks are roughly the same size too!
