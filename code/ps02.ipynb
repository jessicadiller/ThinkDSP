## 6) Apply Bartlett's method to your natural noise source and see if it gives a clearer picture of the relationship between frequency and power.

```python
wave = thinkdsp.read_wave('202018__ryancacophony__singing-bell-hit-1.wav')
wave.ys *= numpy.bartlett(len(wave.ys))
spectrum = wave.make_spectrum()
spectrum.plot_power(low=1, linewidth=1, alpha=0.5)
thinkplot.config(xlabel='frequency (Hz)',
                 ylabel='power density',
                 xscale='log',
                 yscale='log',
                 legend=False)
```

### The central area of the noise is more dense after the Bartlett window has been applied and the spikes outside the central area are either thinner/smaller or just don't exist anymore.
